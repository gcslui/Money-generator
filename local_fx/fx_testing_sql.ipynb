```python
class DatabaseConnection(object):
    def __init__(self, file_name):
        self.file_name = file_name
      
    def __enter__(self):
        self.db = sqlite3.connect(self.file_name)
        return self.db
  
    def __exit__(self, exception_type, exception_value, tb):
        self.db.commit()
        self.db.close()
```

```python
def db_show_all():
    with DatabaseConnection('../local_ms/database_yfinance.db') as db:
        cur = db.cursor()
        cur.execute("SELECT rowid, * FROM {}".format(DB_TABLE_PRIMARY))
        fulltable = cur.fetchall()

        for elem in fulltable:
            print(elem)
    return

def db_add_row(first_name, last_name, email):
    with DatabaseConnection('../local_ms/database_yfinance.db') as db:
        cur = db.cursor()
        cur.execute("INSERT INTO yfinance VALUES (?,?,?)",
                   (first_name, last_name, email))
        fulltable = cur.fetchall()

        for elem in fulltable:
            print(elem)
    return
```

```python
with DatabaseConnection('tickers_yf.db') as db:
    for ticker in ['MSFT', 'AAPL', 'HUT', 'HUT.TO', 'SPY', 'CADUSD=X', 'BTC-USD', 'ETH-USD', 'ETHX-U.TO']:
        tick = yf.Ticker(ticker)

        hist = tick.history(interval='1d',period='max')
        hist['Ticker'] = tick.ticker
        hist.to_sql('daily_data', db, if_exists='append')

        tstart = datetime.datetime.now() - datetime.timedelta(days=7)
        hist = tick.history(interval='1m', start=tstart)
        hist['Ticker'] = tick.ticker
        hist.to_sql('minutely_data', db, if_exists='append')
```

```python
with DatabaseConnection('tickers_yf.db') as db:
    ticker = "\'MSFT\'"
    cur = db.execute('SELECT * FROM daily_data WHERE ticker='+ticker)
    r = cur.fetchall()
    print(r)
```
